```python
import pandas as pd

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

print(df.shape)
print(df.columns)
df.head()
```

```python
df["Sleep Disorder"] = df["Sleep Disorder"].fillna("None")
df["Sleep Disorder"].value_counts()
```

```python
bp = df["Blood Pressure"].str.split("/", expand=True).astype(int)
df["Systolic"] = bp[0]
df["Diastolic"] = bp[1]

df = df.drop(columns=["Blood Pressure"])
df[["Systolic","Diastolic"]].describe()
```

```python
X = df.drop(columns=["Sleep Disorder", "Person ID"])
y = df["Sleep Disorder"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_features = [
    "Gender",
    "Occupation",
    "BMI Category",
    "Physical Activity Level"
]

numerical_features = [
    "Age",
    "Sleep Duration",
    "Quality of Sleep",
    "Stress Level",
    "Heart Rate",
    "Daily Steps",
    "Systolic",
    "Diastolic"
]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

y_train.value_counts(normalize=True)
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
models = {
    "C4.5 (Decision Tree)": DecisionTreeClassifier(criterion="entropy", random_state=42),
    "kNN": KNeighborsClassifier(n_neighbors=5),
    "Naive Bayes": GaussianNB(),
    "Random Forest": RandomForestClassifier(n_estimators=200, random_state=42),
    "AdaBoost": AdaBoostClassifier(n_estimators=200, random_state=42)
}
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report
import pandas as pd
```

```python
results = []

for name, model in models.items():
    pipe = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("smote", smote),
        ("classifier", model)
    ])
    
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Macro F1": f1_score(y_test, y_pred, average="macro")
    })

results_df = pd.DataFrame(results).sort_values(by="Macro F1", ascending=False)
results_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

best_model_name = results_df.iloc[0]["Model"]
best_model = models[best_model_name]

pipe = Pipeline(steps=[
    ("preprocess", preprocessor),
    ("smote", smote),
    ("classifier", best_model)
])

pipe.fit(X_train, y_train)
y_pred = pipe.predict(X_test)

cm = confusion_matrix(y_test, y_pred, labels=["None", "Insomnia", "Sleep Apnea"])

sns.heatmap(cm, annot=True, fmt="d",
            xticklabels=["None", "Insomnia", "Sleep Apnea"],
            yticklabels=["None", "Insomnia", "Sleep Apnea"])
plt.title(best_model_name)
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
import numpy as np
```

```python
cv = StratifiedKFold(
    n_splits=10,
    shuffle=True,
    random_state=42
)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_features),
        ("cat", OneHotEncoder(handle_unknown="ignore", sparse_output=False), categorical_features)
    ]
)

```

```python
cv_results = []

for name, model in models.items():
    pipe = Pipeline(steps=[
        ("preprocess", preprocessor),
        ("smote", smote),
        ("classifier", model)
    ])
    
    acc_scores = cross_val_score(
        pipe, X, y,
        cv=cv,
        scoring="accuracy"
    )
    
    f1_scores = cross_val_score(
        pipe, X, y,
        cv=cv,
        scoring="f1_macro"
    )
    
    cv_results.append({
        "Model": name,
        "CV Accuracy (mean)": acc_scores.mean(),
        "CV Macro F1 (mean)": f1_scores.mean(),
        "CV Macro F1 (std)": f1_scores.std()
    })
```

```python
cv_results_df = pd.DataFrame(cv_results).sort_values(
    by="CV Macro F1 (mean)", ascending=False
)
cv_results_df
```

```python
import pandas as pd

df = pd.read_csv("Sleep_health_and_lifestyle_dataset.csv")

# Target cleaning 
df["Sleep Disorder"] = df["Sleep Disorder"].fillna("None")

# Blood Pressure -> Systolic/Diastolic
bp = df["Blood Pressure"].str.split("/", expand=True).astype(int)
df["Systolic"] = bp[0]
df["Diastolic"] = bp[1]
df = df.drop(columns=["Blood Pressure"])

features = ["Age","Sleep Duration","Quality of Sleep","Stress Level",
            "Heart Rate","Daily Steps","Systolic","Diastolic"]

X = df[features].copy()
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
import os
os.environ["OMP_NUM_THREADS"] = "2"
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertias = []
K_range = range(2, 11)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.figure()
plt.plot(list(K_range), inertias, marker="o")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia (SSE)")
plt.title("Elbow Method for K-Means")
plt.show()
```

```python
k_opt = 3  # ΒΑΛΕ αυτό που βλέπεις από elbow
kmeans = KMeans(n_clusters=k_opt, random_state=42, n_init=10)
labels_km = kmeans.fit_predict(X_scaled)

df["cluster_kmeans"] = labels_km
df["cluster_kmeans"].value_counts()
```

```python
cluster_profile_km = df.groupby("cluster_kmeans")[features].mean().round(2)
cluster_profile_km
```

```python
from sklearn.metrics import silhouette_score
sil_km = silhouette_score(X_scaled, labels_km)
sil_km

```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

Z = linkage(X_scaled, method="ward")  

plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode="level", p=5)  
plt.title("Hierarchical Clustering Dendrogram (Ward)")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

k_hier = 3  # βάσει δενδρογράμματος
agg = AgglomerativeClustering(n_clusters=k_hier, linkage="ward")
labels_agg = agg.fit_predict(X_scaled)

df["cluster_hier"] = labels_agg
df["cluster_hier"].value_counts()
```

```python
cluster_profile_hier = df.groupby("cluster_hier")[features].mean().round(2)
cluster_profile_hier
```

```python
sil_hier = silhouette_score(X_scaled, labels_agg)
sil_hier
```

```python
import numpy as np
from sklearn.neighbors import NearestNeighbors

min_samples = 16
nbrs = NearestNeighbors(n_neighbors=min_samples).fit(X_scaled)
distances, indices = nbrs.kneighbors(X_scaled)

k_distances = np.sort(distances[:, -1]) 

plt.figure()
plt.plot(k_distances)
plt.xlabel("Points sorted by distance")
plt.ylabel(f"{min_samples}-NN distance")
plt.title("k-distance Graph for DBSCAN (choose eps at the elbow)")
plt.show()
```

```python
from sklearn.cluster import DBSCAN

eps_opt = 0.9  
db = DBSCAN(eps=eps_opt, min_samples=min_samples)
labels_db = db.fit_predict(X_scaled)

df["cluster_dbscan"] = labels_db
df["cluster_dbscan"].value_counts()
```

```python
df_db = df[df["cluster_dbscan"] != -1].copy()

cluster_profile_db = df_db.groupby("cluster_dbscan")[features].mean().round(2)
cluster_profile_db
```

```python
mask = df["cluster_dbscan"] != -1

sil_db = silhouette_score(
    X_scaled[mask],
    df.loc[mask, "cluster_dbscan"]
)

sil_db
```
